    <i>Recall that the coefficient estimate $\hat{\beta}$ for the linear regression of $Y$ onto $X$ without an intercept is given by [equation] (3.38). Under what circumstance is the coefficient estimate for the regression of $X$ onto $Y$ the same as the coefficient estimate for the regression of $Y$ onto $X$?</i>
</blockquote>

<p>Recall that from <a href="../Exercise11/3_11.ipynb">Exercise 3.11</a> that equation (3.38) is given by
$$
     \hat{\beta} = \frac{\sum_{i=1}^n x_i y_i}{\sum_{i'=1}^n x_{i'}^2} \,.
$$
We see that the coefficient estimate for the regression of $X$ onto $Y$ is the same as the coefficient estimate for the regression of $Y$ onto $X$ when
$$
\frac{\sum_{i=1}^n x_i y_i}{\sum_{i'=1}^n x_{i'}^2} \stackrel{!}{=} \frac{\sum_{i=1}^n x_i y_i}{\sum_{i'=1}^n y_{i'}^2} \implies \sum_{i=1}^n x_{i}^2 = \sum_{i=1}^n y_{i}^2 \,.
$$</p>

<h3>Exercise 3.12.3</h3>
<blockquote>
    <i>Generate an example in R with $n = 100$ observations in which the coefficient estimate for the regression of $X$ onto $Y$ is <i>the same as</i> the coefficient estimate for the regression of $Y$ onto $X$.</i>
</blockquote>

<p>As discussed above, in this case we must generate two distributions $X$ and $Y$ such that $\sum_{i=1}^n x_{i}^2 = \sum_{i=1}^n y_{i}^2$. To achieve this, we can take any two distributions, calculate the sum of squares of their observations and then calculate the proportion between the sum of squares
$$
\mathrm{proportion} = \frac{\sum_{i=1}^n x_{i}^2}{\sum_{i=1}^n y_{i}^2} \,.
$$
Subsequently, we can rescale one of the original two distributions, for instance
$$
y_i' = \mathrm{sgn} (y_i) * \sqrt{\mathrm{proportion} * y_i^2} \,\,\,\, \forall \, i = 1, 2, \ldots, n \,,
$$
where $\mathrm{sgn}$ denotes the <a href="https://en.wikipedia.org/wiki/Sign_(mathematics)#Sign_functions">sign function</a>. We now see that
$$
        \sum_i^n y'{}_i^2 =  \mathrm{proportion} \sum_i^n y_i^2 = \left( \frac{\sum_{i=1}^n x_{i}^2}{\sum_{i=1}^n y_{i}^2}\right) \sum_{i=1}^n y_i^2 = \sum_{i=1}^n x_{i}^2 \,,
$$
as desired.
</p>


```python
np.random.seed(42)
x = np.random.normal(size=100, loc=0, scale=1)
y = 3*x + np.random.normal(size=100, loc=0, scale=0.8)
# compute the sign function
sgn = np.where(y < 0, -1, 1)
# find the proportion of the sum of x^2 and y^2
proportion = np.sum(x**2) / np.sum(y**2)
# transform the distribution of y so that its
# sum of squares is equal to the sum of squares 
# of the x distribution
y = sgn * (proportion * y**2)**0.5
# assert that the difference between the sum of the squares 
# of the x array elements is equal to the sum of the squares 
# of the y array elements.
assert abs(np.sum(x**2) - np.sum(y**2)) <= 1e-10
df_x = pd.DataFrame({'x': x})
df_y = pd.DataFrame({'y': y})
fitted = sm.OLS(df_y, df_x).fit()
fitted.summary()
createLinearRegressionPlot(df_x, df_y, fitted_model=fitted)
fitted = sm.OLS(df_x, df_y).fit()
fitted.summary()
createLinearRegressionPlot(df_y, df_x, fitted_model=fitted)
```
