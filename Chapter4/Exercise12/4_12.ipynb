<h3>Exercise 4.12.1</h3>
<blockquote>
    <i>Write a function, <code>Power()</code>, that prints out the result of raising $2$ to the $3$rd power. In other words, your function should compute $2^3$ and print out the results. <br><i>Hint: Recall that $x^a$ raises $x$ to the power $a$. Use the <code>print()</code> function to output the result.</i></i>
</blockquote>

```python
def power():
    print(f'2^3 = {2**3}')
    
power()
```

<h3>Exercise 4.12.2</h3>
<blockquote>
    <i>Create a new function, <code>Power2()</code>, that allows you to pass any
two numbers, $x$ and $a$, and prints out the value of $x^a$. You can
do this by beginning your function with the line <br><code>> Power2=function(x,a) {</code><br>You should be able to call your function by entering, for instance,<br><code>> Power2(3,8)</code><br>on the command line. This should output the value of $3^8$, namely, $6,561$.</i>
</blockquote>

```python
def power2(base, exponent):
    print(f'{base}^{exponent} = {base**exponent}')
    
power2(3, 8)
```

<h3>Exercise 4.12.3</h3>
<blockquote>
    <i>Using the <code>Power2()</code> function that you just wrote, compute $10^3$, $8^{17}$, and $131^3$.</i>
</blockquote>

```python
power2(10, 3)
power2(8, 17)
power2(131, 3)
```

<h3>Exercise 4.12.4</h3>
<blockquote>
    <i>Now create a new function, <code>Power3()</code>, that actually returns the result $x^a$ as an R object, rather than simply printing it to the screen. That is, if you store the value $x^a$ in an object called <code>result</code> within your function, then you can simply <code>return()</code> this return() result, using the following line:<br><code>return(result)</code><br>The line above should be the last line in your function, before the <code>}</code> symbol.</i>
</blockquote>

```python
def power3(base, exponent):
    return base**exponent

result = power3(3, 8)
print(result)
```

<h3>Exercise 4.12.5</h3>
<blockquote>
    <i>Now using the <code>Power3()</code> function, create a plot of $f(x)=x^2$. The $x$-axis should display a range of integers from $1$ to $10$, and the $y$-axis should display $x^2$. Label the axes appropriately, and use an appropriate title for the figure. Consider displaying either the $x$-axis, the $y$-axis, or both on the log-scale. You can do this by using <code>log=‘‘x’’</code>, <code>log=‘‘y’’</code>, or <code>log=‘‘xy’’</code> as arguments to the <code>plot()</code> function.</i>
</blockquote>

```python
x = np.arange(1, 11)
y = power3(x, 2)
fig, ax = plt.subplots(constrained_layout=True, figsize=(9, 6))
_ = ax.plot(x, y, '-', linewidth=2)
_ = ax.set_xlabel(r'$x$')
_ = ax.set_ylabel(r'$x^2$')
_ = ax.set_title(r'$f(x) = x^2$')
_ = ax.grid(True)

fig, ax = plt.subplots(constrained_layout=True, figsize=(9, 6))
_ = ax.plot(x, np.log(y), '-', linewidth=2)
_ = ax.set_xlabel(r'$x$')
_ = ax.set_ylabel(r'$\ln(x^2)$')
_ = ax.set_title(r'$f(x) = \ln(x^2)$')
_ = ax.grid(True)
```

<h3>Exercise 4.12.6</h3>
<blockquote>
    <i>Create a function, <code>PlotPower()</code>, that allows you to create a plot of $x$ against $x^a$ for a fixed $a$ and for a range of values of $x$. For instance, if you call<br><code>> PlotPower(1:10,3)</code><br>then a plot should be created with an $x$-axis taking on values $1, 2, \ldots, 10$, and a $y$-axis taking on values $1^3, 2^3, \ldots, 10^3$.</i>
</blockquote>

```python
def plotPower(x_begin, x_end, exponent):
    x = np.arange(x_begin, x_end + 1)
    y = x**exponent
    fig, ax = plt.subplots(constrained_layout=True, figsize=(9, 6))
    _ = ax.plot(x, y, '-', linewidth=2)
    _ = ax.set_xlabel(r'$x$')
    _ = ax.set_ylabel(f'x^{exponent}')
    _ = ax.set_title(f'f(x) = x^{exponent}')
    _ = ax.grid(True)

plotPower(1, 10, 3)    
```
